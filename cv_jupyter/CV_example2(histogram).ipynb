# 영상으로부터 히스토그램을 구하고 표현하기

```
import matplotlib.pyplot as plt
%matplotlib inline   
#%matplotlib inline 의 역할은 notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것
import numpy as np
import cv2

from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
img = x_train[0]
height, width = img.shape
print(height, width)
```

주어진 영상으로부터 히스토그램 및 정규화된 히스토그램을 구하기 위한 프로그램

```
img = x_train[0]

# Histogram
hist, bins = np.histogram(
    img, bins=10, range=(np.min(img), np.max(img))
)

# Normalised Histogram
hist_norm, bins_norm = np.histogram(
    img, bins=10, range=(np.min(img), np.max(img)), normed=True
)

# Visualisation 
plt.figure(figsize=(16,8))
# Subplot 1
plt.subplot(1,2,1)
plt.bar(bins[:-1], hist, width=10)
plt.title("Histogram")

# Subplot 2
plt.subplot(1,2,2)
plt.bar(bins_norm[:-1], hist_norm, width=5)
plt.title("Normalised Histogram")

# Show Plot
plt.show()
```

```
def histogram(img):
    #bins = np.zeros(256, np.int32)
    bins = np.zeros(16, np.int32)
    height, width = img.shape
    for i in range(0, height):
        for j in range(0, width):
            # bins[img[i][j]] += 1
            bins[int(img[i][j]/16)] += 1
    
    bins = bins/(width*height)  #normalised histogram

    return bins

bins =  histogram(img)
plt.plot(bins)
# plt.xlim([0,256])
plt.xlim([0,15])

plt.show()
```

Histogram Stretching and Histogram Equalisation

```
equalised_img = cv2.equalizeHist(img) # opencv를 이용하는 방법의 예

plt.figure(figsize=(14,7))
plt.subplot(121)
plt.imshow(img, cmap="gray")

plt.subplot(122)
plt.imshow(equalised_img, cmap='gray')

plt.show()
```
